```python
import numpy as np
import IPython.display as idisplay
import matplotlib.pyplot as plt
from scipy.io.wavfile import read
```

```python
# sample_rate, waveform = read("audio.wav")
# waveform = np.transpose(waveform)
sample_rate = 44100
waveform = np.sin(np.linspace(0,6,sample_rate*6) * 2 * np.pi * 440) * 1024 # 440 hz sine wave
waveform = np.reshape(waveform, (2, -1))
channels, og_num_samples = waveform.shape
print(waveform.shape, sample_rate)
idisplay.display(idisplay.Audio(waveform, rate=sample_rate))
plt.plot(waveform[0, 100000:105000])
```

```python
window_size = 1000
pitch_up_factor = 2

analysis_hop_length = window_size // pitch_up_factor

synthesis_hop_length = window_size

output_waveform = np.ndarray((channels, og_num_samples*pitch_up_factor))
output_location = 0

for i in range(0, og_num_samples - window_size, analysis_hop_length):
    window = waveform[:, i:i+window_size]
    output_waveform[:, output_location:output_location+window_size] = window
    output_location += synthesis_hop_length

idisplay.display(idisplay.Audio(output_waveform, rate=sample_rate*pitch_up_factor))
plt.plot(output_waveform[0, 100000*pitch_up_factor:105000*pitch_up_factor])
```
